```python
# Importing the needed libraries for the code
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Step 2: Establish the iris variable dataset
iris = sns.load_dataset('iris')
# Show the dataset
iris.head()
```

```python
# Step 3: Remove the column 'species' within the dataset
iris = iris.drop(labels = 'species', axis = 1)
# Show the dataset after action
iris.head()
```

```python
# Step 4: Establish additional columns to the dataset

# Add the 'sepal_sum' column
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
# Add the 'petal_sum' column
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']
# Show the dataset after action
iris.head()
```

```python
# Step 5: Calculate the mean,standard deviation, and min and max values of each column

# Establish the column names next to each row of values
names = ['sepal_length', 'sepal_width', 'sepal_sum', 'petal_length', 'petal_width', 'petal_sum']
# Establish the variable for the mean values
mean = [np.mean(iris['sepal_length']),
        np.mean(iris['sepal_width']),
        np.mean(iris['sepal_sum']),
        np.mean(iris['petal_length']),
        np.mean(iris['petal_width']),
        np.mean(iris['petal_sum'])]
# Establish the variable for the std values
std = [np.std(iris['sepal_length']),
        np.std(iris['sepal_width']),
        np.std(iris['sepal_sum']),
        np.std(iris['petal_length']),
        np.std(iris['petal_width']),
        np.std(iris['petal_sum'])]
# Establish the variable for the min values
min = [np.min(iris['sepal_length']),
        np.min(iris['sepal_width']),
        np.min(iris['sepal_sum']),
        np.min(iris['petal_length']),
        np.min(iris['petal_width']),
        np.min(iris['petal_sum'])]
# Establish the variable for the max values
max = [np.max(iris['sepal_length']),
        np.max(iris['sepal_width']),
        np.max(iris['sepal_sum']),
        np.max(iris['petal_length']),
        np.max(iris['petal_width']),
        np.max(iris['petal_sum'])]

# Establish dictionary of mean values
summary_dict = {'names': names, 'mean': mean, 'std': std, 'min': min, 'max': max}
# Turn values into a pandas dataframe
summary_df = pd.DataFrame(data = summary_dict)
summary_df
```
